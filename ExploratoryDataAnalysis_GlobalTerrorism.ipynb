# Creating new dataframe from given dataset for EDA

```python
#Selecting Useful columns and creating new dataframe as data
data = df[['Year', 'Month','Day','Country','State','Region','City','latitude','longitude','Attacktype','Kill','Wound','target1',
          'summary','Group', 'Targettype', 'Weapon', 'motive']]
```

```python
# check null values
data.isnull().sum()
```

```python
# replace null values with 0
data['Wound'] = data['Wound'].fillna(0)
data['Kill'] = data['Kill'].fillna(0)
```

# Exploratory Data Analysis

## Number of terrorist Attacks in each year.

```python
# Attacks in year
year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()

plt.figure(figsize= (20, 10))
sns.barplot(x= year, y = years_count, palette= "tab20")

plt.xticks(rotation = 90)
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Number of Attacks Each Year', fontsize = 20)

plt.show()

attacks=data["Year"].value_counts(dropna=False).sort_index() 
attacks=pd.DataFrame(attacks).reset_index().rename(columns={"index":"Year","Year":"Attacks"}).set_index("Year")
attacks.tail(7)
```

There is tremendous growth in terrorist attacks after 2008.

Least terrorist attack at 1971.

## Number of people killed in terrorist attack each year.

```python
kills_year=(data.groupby("Year").sum()["Kill"])

#Plotting 
plt.figure(figsize=(16,5))
kills_year.plot(kind="bar", color="red")

plt.title('People Killed Due to Terrorist Attack', fontsize = 20)
plt.xlabel('Year', fontsize= 15)
plt.ylabel('Number of Killed People', fontsize = 15)
plt.legend(loc='upper left', fontsize=15)

plt.show()

#kills_year=pd.DataFrame(kills_year)
#kills_year
```

Most people died in terrorist attack is in 2014.

## Number of people wounded in terrorist attack each year.

```python
Wound_year=(data.groupby("Year").sum()["Wound"])


#Plotting 
plt.figure(figsize=(16,5))
Wound_year.plot(kind="bar", color="red")
plt.title('People Wounded Due to Terrorist Attack', fontsize = 20)
plt.xlabel('Year', fontsize= 15)
plt.ylabel('Number of Wounded People', fontsize = 15)
plt.legend(loc='upper left', fontsize=15)
plt.show()
```

Most people wounded in terrorist attack is in 2015

## Types of terrorist attacks

```python
# Types of terrorist attacks

plt.figure(figsize=(16,10))
plt.xticks(rotation = 90)
sns.countplot(x="Attacktype", data=data)
plt.xlabel("Attacktype",fontsize=16)
plt.xticks(fontsize=13)
plt.tight_layout()
```

## Hotspots for terrorist attacks

```python
# Hotspots for terrorist attacks
plt.figure(figsize=(15, 7))
sns.countplot(data.Targettype, order= data.Targettype.value_counts().index, palette="ocean")

plt.xticks(rotation= 90)
plt.xlabel('Type of Attack', fontsize= 20)
plt.ylabel('Count', fontsize= 20)
plt.title('Hotspots for terrorist attacks', fontsize= 25)

plt.show()
```

Terrorist preferred to attack soft targets like private citizens and private properties.

Iraq is the most affected country by the terrorist attacks.

## Region Wise Analysis

```python
regionAttack=pd.crosstab(df.Year,df.Region)

regionAttack.plot(kind="area", stacked=False, alpha=0.5,figsize=(16,9))
plt.title("Region wise attacks",fontsize=20)
plt.xlabel("Years",fontsize=20)
plt.ylabel("Number of Attacks",fontsize=20)
plt.show()
```

Middle east and north Africa are most affected region in terrorist attacks.

## Analysis of Terrorist groups

### Top10 terrorist groups

```python
#plotting top10 terrorist groups
terroristGroup=data["Group"].value_counts().drop('Unknown')[0:10]

plt.figure(figsize=(16,7))
sns.barplot(terroristGroup.index,terroristGroup)
plt.title("Group Wise Attacks",fontsize=16)
plt.xlabel("Terrorist Groups",fontsize=16)
plt.xticks(rotation= 80,fontsize=10)
plt.ylabel("Number of Attacks",fontsize=16)
plt.show()
```

**Creating new dataframe based on 'Group', 'Country','Kill','Wound'.**

```python
#Creating new dataframe
newdata= data[['Group', 'Country','Kill','Wound']]
newdata=newdata.groupby(["Group","Country"], axis= 0).sum().sort_values('Kill', ascending= False).drop('Unknown').reset_index().head(10)
newdata
```

## Top 10 Terrorist groups as per Kills

```python
# Top 10 Terrorist groups as per Kills
killbyGroups=(newdata.groupby("Group").sum()["Kill"]).sort_values(ascending=False)[:10]
killbyGroups

#Plotting 
plt.figure(figsize=(16,5))
killbyGroups.plot(kind="bar", color="red")

plt.title('People killed By a Terrorist Groups', fontsize = 20)
plt.xlabel('Terrorist Group', fontsize= 15)
plt.ylabel('Number of killed People by a Terrorist Group', fontsize = 10)
plt.legend(loc=0, fontsize=15)
plt.show()

killbyGroups
```

Islamic State of Iraq and the Levant (ISIL) killed 37941 people.

## Top 10 Terrorist groups as per wounds

```python
woundbyGroups=(newdata.groupby("Group").sum()["Wound"]).sort_values(ascending=False)[:11]

#Plotting 
plt.figure(figsize=(16,5))
woundbyGroups.plot(kind="bar", color="red")

plt.title('People Wounded By a Terrorist Groups', fontsize = 20)
plt.xlabel('Terrorist Group', fontsize= 15)
plt.ylabel('Number of Wounded People by a Terrorist Group', fontsize = 10)
plt.legend(loc=0, fontsize=15)
plt.show()

woundbyGroups
```

Taliban wounded 27871 people in terrorist attacks.

## Finding which terrorist group active in which country.

```python
groupandCountries=data[["Group","Country"]].value_counts().drop("Unknown")

groupandCountries.head(10).plot(kind="bar",color="red",figsize=(16,7))
plt.title("Countries with most attacks by a particular Terrorist Group",fontsize=13)
plt.xlabel("(Terrorist Group,Country)",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Attacks",fontsize=13)
plt.show()

groupandCountries.head(10)
```

### Plotting top10 countries which are attacked by particular terrorist group and killed people in attack.

```python
KillgroupandCountries=data[["Group","Country","Kill"]].groupby(["Group","Country"],axis=0).sum().sort_values(by="Kill",ascending=False).drop("Unknown").head(10)

KillgroupandCountries.plot(kind="bar",color="red",figsize=(15,6))
plt.title("Countries with most Killed people by a particular Terrorist Group",fontsize=13)
plt.xlabel("(Terrorist Group,Country)",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()

KillgroupandCountries
```

### Plotting top10 countries which are attacked by particular terrorist group and wounded people in attack.

```python
WoundgroupandCountries=data[["Group","Country","Wound"]].groupby(["Group","Country"],axis=0).sum().sort_values(by="Wound",ascending=False).drop("Unknown").head(10)

WoundgroupandCountries.plot(kind="bar",color="red",figsize=(15,6))
plt.title("Countries with most Wounded people by a particular Terrorist Group",fontsize=13)
plt.xlabel("(Terrorist Group,Country)",fontsize=13)
plt.xticks(fontsize=12)
plt.ylabel("Number of Casualties",fontsize=13)
plt.show()

WoundgroupandCountries
```
